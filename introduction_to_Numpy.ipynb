## Dot product 

```python
np.random.seed()

mat1 = np.random.randint(10, size = (5,4))
mat2 = np.random.randint(10, size = (5,4))

mat1
```

```python
mat2
```

```python
mat1.shape, mat2.shape
```

```python
# Element-Wise Multiplication 
mat1 * mat2
```

```python
# Dot product
np.dot(mat1, mat2)
```

```python
# Transpose mat1
mat1 = mat1.T
```

```python
mat1
```

```python
mat1.shape
```

```python
mat3 = np.dot(mat1, mat2)
```

```python
mat3
```

```python
mat3.shape
```
